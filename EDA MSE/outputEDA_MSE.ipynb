```python
import os

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

import re
```

```python
def draw_box_plot(df, epoch):
    label_list = [row['label'] for i, row in df.iterrows()]
    pred_list = [row['pred'] for i, row in df.iterrows()]
    plt.boxplot([label_list, pred_list])
    plt.xticks([1, 2], ['label', 'pred'])
    plt.title(f"Boxplot for target label when epoch {epoch}")
    plt.show()
    
    
def matching_plot(df):
    # Matching Plot
    
    _x = np.arange(len(df['label']))
    index = df['label'].sort_values().index
    plt.figure(figsize=(20,8))
    plt.plot(_x, df['pred'][index], 'rx', label='prediction')
    plt.plot(_x, df['label'][index], 'g+', label='label')
    #plt.plot(_x, abs(df['pred'][index]- df['label'][index]), 'yx', label='differ')
    plt.legend()
    plt.show()
    
def show_match(df, header):
    _x = np.arange(len(df['label']))
    diff = abs(df['pred']- df['label'])
    index = diff.sort_values().index
    
    after_s1 = [] 
    after_s1_pad = []
    for s in df['s1'][index]: 
        after_s1_pad.append(s.count('[PAD]'))
        after_s1.append(re.sub('\[PAD\]', "", s))
        
    after_s2 = [] 
    after_s2_pad = []
    for s in df['s2'][index]: 
        after_s2_pad.append(s.count('[PAD]'))
        after_s2.append(re.sub('\[PAD\]', "", s))
               
    show = pd.DataFrame({#'idx':index, 
                         'label': df['label'][index].round(2),
                         'diff':diff[index].round(2),
                         's1_pad': after_s1_pad,
                         's2_pad': after_s2_pad,
                         's1': after_s1,
                         's2': after_s2,
                         })
    show.to_csv(f'{header}_result_show.csv')
```

```python
epoch = len(os.listdir('output'))
for e in range(epoch):
    if e == 49 or e % 5 == 0:
        data = pd.read_csv(f'output/train_{e}.csv')
        show_match(data, "train")
        print(e, data.describe())
        draw_box_plot(data, e)
        matching_plot(data)
```

```python
epoch = len(os.listdir('output'))
for e in range(epoch):
    if e == epoch-1 or e % 5 == 0:
        data = pd.read_csv(f'output/val_{e}.csv')
        show_match(data, "val")
        print(e, data.describe())
        draw_box_plot(data, e)
        matching_plot(data)
```
